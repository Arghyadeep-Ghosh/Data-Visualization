```python
# Q5) Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate a synthetic dataset with correlated features
np.random.seed(42)  # For reproducibility
n_samples = 100

# Generate random data
x = np.random.rand(n_samples)
# Create correlated features
y = 2 * x + np.random.normal(0, 0.1, n_samples)  # y is correlated with x
z = -x + np.random.normal(0, 0.1, n_samples)  # z is negatively correlated with x

# Create a DataFrame
data = pd.DataFrame({'X': x, 'Y': y, 'Z': z})

# Calculate the correlation matrix
correlation_matrix = data.corr()

# Apply Seaborn theme for styling
sns.set_theme()

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar_kws={"shrink": .8})

# Add title
plt.title("Correlation Matrix Heatmap")

# Display the plot
plt.show()
```

```python
'''
 Q4) Using the given x and y data, generate a 3D surface plot to visualize the function z=sin(sqrt(x**2+y**2))
 x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)   '''

import numpy as np
import pandas as pd
import plotly.graph_objects as go

# Generate the x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# Calculate z
z = np.sin(np.sqrt(x**2 + y**2))

# Prepare the data for the DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create a 3D surface plot
fig = go.Figure(data=[go.Surface(z=df['Z'].values.reshape(100, 100),
                                   x=df['X'].values.reshape(100, 100),
                                   y=df['Y'].values.reshape(100, 100))])

# Update layout
fig.update_layout(title='3D Surface Plot of z = sin(sqrt(x^2 + y^2))',
                  scene=dict(
                      xaxis_title='X',
                      yaxis_title='Y',
                      zaxis_title='Z'),
                  autosize=True)

# Show the plot
fig.show()
```

```python
# Q1) Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Prepare the output to display in a Jupyter notebook
output_notebook()

# Generate x values from 0 to 10
x = np.linspace(0, 10, 100)
# Calculate y values as the sine of x
y = np.sin(x)

# Create a new Bokeh plot
p = figure(title='Sine Wave', x_axis_label='X', y_axis_label='Sine(X)')

# Add a line renderer for the sine wave
p.line(x, y, legend_label='Sine Wave', line_width=2, line_color='blue')

# Show the plot
show(p)
```

```python
'''
Q4)  Create a Bokeh histogram to visualize the distribution of the given data.

data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)   '''

import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Prepare the output to display in a Jupyter notebook
output_notebook()

# Generate random data
data_hist = np.random.randn(1000)

# Calculate the histogram
hist, edges = np.histogram(data_hist, bins=30)

# Create a new Bokeh plot
p = figure(title='Histogram of Random Data',
           x_axis_label='Value',
           y_axis_label='Frequency',
           tools='save',
           toolbar_location=None)

# Add a quad renderer for the histogram bars
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color='blue', line_color='black')

# Show the plot
show(p)


```

```python
'''
Q5)  Create a Bokeh heatmap using the provided dataset.

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)      '''

import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar, LinearColorMapper
from bokeh.models import ColumnDataSource

# Prepare the output to display in a Jupyter notebook
output_notebook()

# Generate random heatmap data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Flatten the data for the heatmap
data = data_heatmap.flatten()
x_pos = xx.flatten()
y_pos = yy.flatten()

# Create a color mapper
mapper = LinearColorMapper(palette="Viridis256", low=data.min(), high=data.max())

# Create a new Bokeh plot
p = figure(title='Heatmap',
           x_axis_label='X-axis',
           y_axis_label='Y-axis',
           tools='',
           width=400,
           height=400)

# Add the rectangles to the plot
source = ColumnDataSource(data=dict(x=x_pos, y=y_pos, value=data))
p.rect(x='x', y='y', width=0.1, height=0.1, source=source,
        fill_color=linear_cmap('value', palette='Viridis256', low=data.min(), high=data.max()),
        line_color=None)

# Add a color bar
color_bar = ColorBar(color_mapper=mapper, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```
